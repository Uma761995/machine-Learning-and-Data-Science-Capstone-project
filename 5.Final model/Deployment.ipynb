```python
import pickle
```

```python
def final_result():
    loaded_model=pickle.load(open("Finalized model_ILPD.sav",'rb'))
    Age=int(input("Enter the Age:"))
    Alkphos=float(input("Enter the Alkphos:"))   
    Sgot=int(input("Enter the Sgot:"))
    result=loaded_model.predict([[Age,Alkphos,Sgot]])
    return result
```
