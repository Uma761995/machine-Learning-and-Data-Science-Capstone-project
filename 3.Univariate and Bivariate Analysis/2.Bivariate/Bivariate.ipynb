```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif['Varaince']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif
```

```python
#ttest_independent variable-different groups and same condition.
#groups must be in categorical and conditions must be in numerical features.
```

```python
from scipy.stats import ttest_ind
Male=dataset[dataset['Gender']=='Male']['Age']
Female=dataset[dataset['Gender']=='Female']['Age']
ttest_ind(Male,Female)
```

```python

def indep_ttest(dataset,group_col,group1,group2,value_col):
    from scipy.stats import ttest_ind
    Male=dataset[dataset[group_col]==group1][value_col]
    Female=dataset[dataset[group_col]==group2][value_col]
    ttest=ttest_ind(Male,Female)
    pvalue=ttest.pvalue
    print('P_Value:',pvalue)
    if(pvalue<0.05):
        print("Reject null hypothesis:There is a significant relationship between given varables")
    else:
        print("Fail to Reject null hypothesis: There is no significant relationship between given varables")
    return ttest
```

```python
indep_ttest(dataset,'Gender','Male','Female','Selector')
```

```python
# There is no similarity between Gender and selector because pvalue is 0.046=0.05.
#if pvalue>=0.05,Fail to reject
```

```python
indep_ttest(dataset,'Gender','Male','Female','Age')
```

```python
#pvalue is 0.17>0.05, so there is no significant relationship between Gender and Age.
```

```python
indep_ttest(dataset,'Gender','Male','Female','TP')
```

```python
#pvalue is 0.03<0.05, so there is significant relationship between Gender and TP.
```

```python
#pvalue is 0<0.05, so there is significant relationship between Gender and Sgpt.
```

```python
#ttest_dependent variable-same group and different condition.
def dep_ttest(dataset,group_col,group,value1,value2):  
    from scipy.stats import ttest_rel
    male=dataset[dataset[group_col]==group][value1]
    male1=dataset[dataset[group_col]==group][value2]
    ttest=ttest_rel(male,male1)
    pvalue=ttest.pvalue
    print("P_Value:",pvalue)
    if(pvalue<0.05):
        print("Reject null hypothesis:There is a significant relationship between given varables")
    else:
        print("Fail to Reject null hypothesis: There is no significant relationship between given varables")
    return ttest
```

```python
#Anova-Analysis of variance tells the significant difference between two variables
#H0-There is no significant difference between given variables
#H1-There is a significant difference between given variables
```

```python
# one way classification-one Independent variable
def anova_oneway(dataset,var1,var2,var3):
    from scipy.stats import stats
    one_way=stats.f_oneway(dataset[var1],dataset[var2],dataset[var3])
    pvalue=one_way.pvalue
    print('one_way_pvalue:',pvalue)
    if(pvalue<0.05):
        print("Reject null hypothesis:There is a significant relationship between given varables")
    else:
        print("Fail to Reject null hypothesis: There is no significant relationship between given varables")
    return one_way
```

```python
anova_oneway(dataset,'Sgpt','Sgot','TP')
```

```python
#H0-fail to reject null hypothesis-There is no significant difference between given variables.
#H1-Reject null hypothesis-There is a significant difference between given variables.
#pvalue=0>0.05, if pvalue<0.05-Reject null hypothesis.
#So, here significant difference between Sgpt,Sgot,TP variables
```

```python
anova_oneway(dataset,'Age','TB','DB')
```

```python
#H0-fail to reject null hypothesis-There is no significant difference between given variables.
#H1-Reject null hypothesis-There is a significant difference between given variables.
#pvalue=0>0.05, if pvalue<0.05-Reject null hypothesis.
#So, here H1 significant difference between Age,TP,DB variables
```

```python
anova_oneway(dataset,'Age','Alkphos','ALB')
```

```python
#pvalue=0>0.05, if pvalue<0.05-Reject null hypothesis.
#So, here H1 significant difference between Age,Alkphos,ALB variables
```

```python
# Anova_Two way classification-independent is categorical and dependent is numerical features

def anova_two_way(dataset,numer,cate1,cate2):
    import statsmodels.api as sm   #creating the table    
    from statsmodels.formula.api import ols   #creating the model
    model=ols(f'{numer}~C({cate1})+C({cate2})+C({cate1}):C({cate2})',data=dataset).fit()
    anova_table = sm.stats.anova_lm(model, type=2)
    return anova_table
```

```python
anova_two_way(dataset,'Age','Gender','Selector')
```

```python
#Age + Gender=pvalue of 0.16>0.05 so H0 there is no significant difference between these two varables
#Age + Selector=pvalue of 0<0.05 so H0 there is significant difference between these two varables
#Gender:Selector=pvalue of 0.08>0.05 so H0 there is no significant difference between these two varables
```

```python
anova_two_way(dataset,'TB','Gender','Selector')
```
