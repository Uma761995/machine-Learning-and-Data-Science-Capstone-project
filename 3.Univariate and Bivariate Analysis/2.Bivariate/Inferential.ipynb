```python
import pandas as pd
```

```python
dataset=pd.read_csv("Preprocessed ILPD.csv")
dataset
```

```python
def quanqual(dataset):
        quan=[]
        qual=[]
        for columnName in dataset.columns:
            if (dataset[columnName].dtype=='O'):
                qual.append(columnName)
            else:
                quan.append(columnName)
        return quan,qual
```

```python
quan,qual=quanqual(dataset)
```

```python
# How many male and female affect liver disorder?

dataset.groupby('Gender')['Selector'].sum()
```

```python
#Who are all above 60 age affected person?

dataset.groupby(dataset['Age']>60)['Gender'].value_counts().reset_index()
```

```python
#Reason for affected person?

dataset.groupby('Selector')[quan].mean()
```
